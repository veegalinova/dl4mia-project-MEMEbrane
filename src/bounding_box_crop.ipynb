```python
import os
import numpy as np
import skimage.io as skio
from skimage.morphology import binary_dilation, disk
from skimage.measure import regionprops, label
from glob import glob
```

```python
parent_folder = "/group/dl4miacourse/projects/membrane/ecad_gfp"
save_folder = "/group/dl4miacourse/projects/membrane/ecad_gfp_cropped"
```

```python
split_folder = "train"
for subfolder in os.listdir(parent_folder):
    print(subfolder)
    subfolder_path = os.path.join(parent_folder, subfolder)
    if not os.path.isdir(subfolder_path):
        continue
    
    for this_file in os.listdir(subfolder_path):
        if "C488" in this_file:
            im_path = os.path.join(subfolder_path, this_file)
        elif "cell_mesh_mask.tif" in this_file:
            mask_path = os.path.join(subfolder_path, this_file)

    print(im_path, mask_path)

    im_name = os.path.splitext(os.path.split(im_path)[1])[0]
    mask_name = os.path.splitext(os.path.split(mask_path)[1])[0]

    img = skio.imread(im_path)
    mask = skio.imread(mask_path)

    for t, [this_slice, this_mask] in enumerate(zip(img, mask)):
        if np.sum(this_mask) == 0:
            continue
        bin_mask = this_mask > 0
        dilated_mask = binary_dilation(bin_mask, disk(3))
        this_region = regionprops(label(dilated_mask))
        if len(this_region) != 1:
            print("weird!")
        
        min_row, min_col, max_row, max_col = this_region[0].bbox

        cropped_mask = this_mask[min_row:max_row, min_col:max_col].astype(np.uint16)
        cropped_img = this_slice[min_row:max_row, min_col:max_col].astype(np.uint16)

        skio.imsave(os.path.join(save_folder, split_folder, "im", f"{subfolder}_{im_name}_{t}.tif"), cropped_img)
        skio.imsave(os.path.join(save_folder, split_folder, "mask", f"{subfolder}_{mask_name}_{t}.tif"), cropped_mask)
```
